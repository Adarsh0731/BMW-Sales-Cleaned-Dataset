# Single Filtration

```python
# Single filtration: Year > 2020
single_filter = df[df["Year"] > 2020]

single_filter.head()
```

```python
# Single filtration: Fuel type is Petrol
single_filter_fuel = df[df["Fuel_Type"] == "Petrol"]

single_filter_fuel.head()
```

# Doublr Filtration

```python
# Double filtration using AND (&)
double_filter = df[
    (df["Fuel_Type"] == "Petrol") &
    (df["Price_USD"] > 50000)
]

double_filter.head()
```

```python
double_filter_region = df[
    (df["Region"] == "Europe") &
    (df["Sales_Volume"] > 5000)
]

double_filter_region.head()
```

# Triple Filtration

```python
# Triple filtration
triple_filter = df[
    (df["Fuel_Type"] == "Diesel") &
    (df["Year"] > 2018) &
    (df["Price_USD"] < 70000)
]

triple_filter.head()
```

```python
triple_filter_model = df[
    (df["Model"].str.contains("X")) &
    (df["Region"] == "Asia") &
    (df["Sales_Volume"] > 4000)
]

triple_filter_model.head()
```

# Single Filtration With Sorting

```python
# Filter cars after 2020 and sort by Price (ascending)
single_sorted = df[df["Year"] > 2020].sort_values(by="Price_USD")

single_sorted.head()
```

# Double Filtration With Sorting

```python
# Petrol cars with price > 50000 sorted by Sales Volume (descending)
double_sorted = df[
    (df["Fuel_Type"] == "Petrol") &
    (df["Price_USD"] > 50000)
].sort_values(by="Sales_Volume", ascending=False)

double_sorted.head()
```

# Triple Filtration With Sorting

```python
# Diesel cars after 2018 with price < 70000 sorted by Mileage
triple_sorted = df[
    (df["Fuel_Type"] == "Diesel") &
    (df["Year"] > 2018) &
    (df["Price_USD"] < 70000)
].sort_values(by="Mileage_KM")

triple_sorted.head()
```

# Using groupby Method

```python
region_avg_price = df.groupby("Region")["Price_USD"].mean().sort_values(ascending=False)
region_avg_price
```

```python
model_avg_mileage = df.groupby("Model")["Mileage_KM"].mean().sort_values()
model_avg_mileage
```

# Making Pivot Table using Columns

```python
pivot_region_price = pd.pivot_table(
    df,
    values="Price_USD",
    index="Region",
    aggfunc="mean"
)

pivot_region_price
```

# **Data Visualization**

# 1. Total Sales Trend Over Years

```python
import matplotlib.pyplot as plt

# Group data by Year and calculate total sales volume
yearly_sales = df.groupby("Year")["Sales_Volume"].sum()

# Line chart to analyze sales trend over time
plt.figure()
plt.plot(yearly_sales.index, yearly_sales.values)
plt.xlabel("Year")
plt.ylabel("Total Sales Volume")
plt.title("BMW Total Sales Trend Over Years")
plt.show()
```

# 2. Average Price by Fuel Type

```python
# Calculate average price for each fuel type
avg_price_fuel = df.groupby("Fuel_Type")["Price_USD"].mean()

# Bar chart to compare pricing across fuel types
plt.figure()
avg_price_fuel.plot(kind="bar")
plt.xlabel("Fuel Type")
plt.ylabel("Average Price (USD)")
plt.title("Average BMW Price by Fuel Type")
plt.show()
```

# 3. Total Sales Volume by Region

```python
# Calculate total sales volume for each region
region_sales = df.groupby("Region")["Sales_Volume"].sum()

# Bar chart to identify top-performing regions
plt.figure()
region_sales.plot(kind="bar")
plt.xlabel("Region")
plt.ylabel("Total Sales Volume")
plt.title("Total BMW Sales Volume by Region")
plt.show()
```

# **4.**Price** vs Mileage Relationship**

```python
# Scatter plot to analyze relationship between mileage and price
plt.figure()
plt.scatter(df["Mileage_KM"], df["Price_USD"])
plt.xlabel("Mileage (KM)")
plt.ylabel("Price (USD)")
plt.title("BMW Price vs Mileage Relationship")
plt.show()
```

# 5. Engine Size Distribution

```python
# Histogram to understand distribution of engine sizes
plt.figure()
plt.hist(df["Engine_Size_L"])
plt.xlabel("Engine Size (Liters)")
plt.ylabel("Frequency")
plt.title("Distribution of BMW Engine Sizes")
plt.show()
```

# 6. Sales Volume Distribution

```python
# Histogram to understand sales volume spread
plt.figure()
plt.hist(df["Sales_Volume"])
plt.xlabel("Sales Volume")
plt.ylabel("Frequency")
plt.title("Distribution of BMW Sales Volume")
plt.show()
```

# 7.Correlation Heatmap (Numerical Columns)

```python
import matplotlib.pyplot as plt
import numpy as np

# Select only numerical columns for correlation
numeric_df = df.select_dtypes(include=[np.number])

# Calculate correlation matrix
corr_matrix = numeric_df.corr()

# Create heatmap using matplotlib
plt.figure()
plt.imshow(corr_matrix)
plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=90)
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)
plt.colorbar()
plt.title("Correlation Heatmap of BMW Numerical Features")
plt.show()
```

# 8. Boxplot: Price Distribution (Outlier Detection)

```python
# Boxplot to detect outliers in car prices
plt.figure()
plt.boxplot(df["Price_USD"])
plt.xlabel("BMW Cars")
plt.ylabel("Price (USD)")
plt.title("Boxplot of BMW Car Prices (Outlier Detection)")
plt.show()
```

```python
# Boxplot to analyze mileage outliers
plt.figure()
plt.boxplot(df["Mileage_KM"])
plt.xlabel("BMW Cars")
plt.ylabel("Mileage (KM)")
plt.title("Boxplot of BMW Mileage Distribution")
plt.show()
```
